```python
class Node:
    def __init__(self, pertanyaan="Tidak ada", ya=None, tidak=None, hasil=None, link=None):
        self.pertanyaan = pertanyaan
        self.ya = ya
        self.tidak = tidak
        self.hasil = hasil
        self.link = link

def menjalankan(node):
    while node:
        if node.hasil is not None:
            print("\nHasil Rekomendasi Skincare:\n")
            print(node.hasil)
            if node.link:
                print(f"Link Produk: {node.link}")
            break
        jawab = input(f"{node.pertanyaan} (ya/tidak): ").strip().lower()
        if jawab == 'ya':
            node = node.ya
        elif jawab == 'tidak':
            node = node.tidak
        else:
            print("Jawaban tidak valid. Harap masukkan 'ya' atau 'tidak'.")

def buat_tree():
    # Node hasil akhir dengan rekomendasi skincare dan link produk
    node_hasil_kering_indoor = Node(
        hasil="Kulit Anda kering dan Anda sering berada di dalam ruangan. Gunakan pelembap intensif serta produk yang menjaga kelembapan kulit, seperti pelembap dari Cetaphil.",
        link="https://id.shp.ee/ziZYrfq"
    node_hasil_kering_iritasi = Node(
        hasil="Kulit Anda kering dan sering mengalami iritasi. Gunakan pelembap dengan bahan anti-inflamasi seperti ceramide atau centella, seperti produk dari Skin tific atau El Formula.",
        link="https://id.shp.ee/r9svAkp"
    )
    node_hasil_kering_kusam = Node(
        hasil="Kulit Anda kering dan terlihat kusam setelah aktivitas. Disarankan menggunakan pelembap dari Cetaphil dan serum yang mengandung vitamin C.",
       link="https://id.shp.ee/ziZYrfq"
        link="https://id.shp.ee/ftmMwMS"
    )
    node_hasil_berminyak_jerawat_ringan = Node(
        hasil="Kulit Anda berminyak dengan jerawat ringan setelah aktivitas. Gunakan produk perawatan kulit yang mengontrol minyak, seperti pelembap dari COSRX dan pembersih dengan asam salisilat.",
        link="https://id.shp.ee/o5NHZsN"
    )
    node_hasil_berminyak_jerawat_berat = Node(
        hasil="Kulit Anda berminyak dan sering mengalami jerawat setelah olahraga berat. Disarankan menggunakan pelembap dari Kiehl's dan produk berbasis niacinamide untuk membantu mengurangi jerawat.",
        link= "https://id.shp.ee/Gj8GnRi"
    )

    # Masalah untuk aktivitas di luar ruangan pada kulit kering
    node_kering_masalah = Node(
        "Apakah Anda sering mengalami iritasi atau kemerahan setelah beraktivitas di luar ruangan?",
        ya=node_hasil_kering_iritasi, tidak=node_hasil_kering_kusam
    )

    # Masalah untuk kulit berminyak dengan aktivitas ringan atau berat
    node_berminyak_jerawat = Node(
        "Apakah Anda mengalami jerawat setelah aktivitas?",
        ya=node_hasil_berminyak_jerawat_berat,
        tidak=node_hasil_berminyak_jerawat_ringan
    )

    # Pertanyaan aktivitas untuk kulit berminyak
    node_berminyak_berat = Node(
        "Apakah Anda sering melakukan olahraga berat?",
        ya=node_berminyak_jerawat,
        tidak=node_hasil_berminyak_jerawat_ringan
    )
    node_berminyak_aktivitas = Node(
        "Apakah aktivitas Anda kebanyakan di dalam ruangan?",
        ya=node_berminyak_berat,
        tidak=node_berminyak_berat
    )

    # Pertanyaan aktivitas untuk kulit kering
    node_kering_outdoor = Node(
        "Apakah aktivitas Anda kebanyakan di luar ruangan?",
        ya=node_kering_masalah,
        tidak=node_hasil_kering_indoor
    )

    # Pertanyaan utama jenis kulit
    node_berminyak = Node(
        "Apakah kulit Anda terasa berminyak di seluruh wajah?",
        ya=node_berminyak_aktivitas,
        tidak=node_kering_outdoor
    )
    node_kering = Node(
        "Apakah kulit Anda terasa kering atau ketat setelah mencuci muka?",
        ya=node_kering_outdoor,
        tidak=node_berminyak
    )

    # Root node
    root = Node(
        "Apakah jenis kulit Anda? (ya untuk kering, tidak untuk berminyak)",
        ya=node_kering,
        tidak=node_berminyak
    )

    return root

if __name__ == "__main__":
    tree = buat_tree()
    menjalankan(tree)
```
